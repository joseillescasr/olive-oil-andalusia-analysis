```python
import pandas as pd

# Ruta al archivo
file_path = '/content/drive/MyDrive/proyecto_aceite_oliva/Olivar Superficie7.12.xlsx'

# Ver todas las pestañas disponibles
xl = pd.ExcelFile(file_path)
print("Pestañas disponibles:")
for i, sheet in enumerate(xl.sheet_names):
    print(f"  {i+1}. {sheet}")
```

```python
# Exploración inicial de las 6 tablas prioritarias
sheets_prioritarias = ['7.12.1.5', '7.12.1.6', '7.12.1.8.1',
                       '7.12.2.3', '7.12.3.1', '7.12.3.2']

for sheet in sheets_prioritarias:
    df = pd.read_excel(file_path, sheet_name=sheet, header=None)
    print(f"\n{'='*50}")
    print(f"PESTAÑA: {sheet}")
    print(f"Dimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")
    print(f"Primeras 5 filas:")
    print(df.head())
```

```python
# Exploramos la tabla de precios históricos en detalle
df_precios = pd.read_excel(file_path, sheet_name='7.12.3.2', header=None)

print(f"Dimensiones: {df_precios.shape}")
print("\nContenido completo:")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
print(df_precios.to_string())
```

```python
# Limpieza tabla 7.12.3.2 - Precios históricos aceite de oliva virgen

# Leemos solo las filas con datos reales (años + precios)
# Nos quedamos con columna 0 (año) y columna 2 (precio calidad superior)
df_precios_clean = pd.read_excel(file_path, sheet_name='7.12.3.2', header=None)

# Extraemos solo filas donde la columna 0 es un año (número entre 2000 y 2023)
df_precios_clean = df_precios_clean[
    pd.to_numeric(df_precios_clean[0], errors='coerce').between(2000, 2023)
].copy()

# Renombramos columnas
df_precios_clean = df_precios_clean[[0, 2, 3, 4]].copy()
df_precios_clean.columns = ['anio', 'precio_calidad_alta', 'precio_calidad_media', 'precio_calidad_baja']

# Convertimos a numérico
for col in ['precio_calidad_alta', 'precio_calidad_media', 'precio_calidad_baja']:
    df_precios_clean[col] = pd.to_numeric(df_precios_clean[col], errors='coerce')

df_precios_clean['anio'] = df_precios_clean['anio'].astype(int)

# Reseteamos índice
df_precios_clean = df_precios_clean.reset_index(drop=True)

print(f"Dimensiones tabla limpia: {df_precios_clean.shape}")
print("\nVista completa:")
print(df_precios_clean.to_string())
```

```python
# Guardamos la tabla limpia como CSV en Google Drive
# Esto es importante: siempre guardamos los datos limpios separados de los crudos
output_path = '/content/drive/MyDrive/proyecto_aceite_oliva/precios_historicos_clean.csv'

df_precios_clean.to_csv(output_path, index=False)
# index=False evita que guarde la columna del índice (0,1,2...) en el CSV — no la necesitamos

print(f"✅ Archivo guardado en: {output_path}")
print(f"Filas: {df_precios_clean.shape[0]} | Columnas: {df_precios_clean.shape[1]}")
```

```python
# Limpieza tabla 7.12.3.1 - Aceite virgen por tipos y provincia 2023

# Leemos la pestaña sin tratar ninguna fila como encabezado
df_31_clean = pd.read_excel(file_path, sheet_name='7.12.3.1', header=None)

# Usamos la fila 4 como encabezado de columnas
df_31_clean.columns = df_31_clean.iloc[4]

# Nos quedamos solo desde la fila 5 en adelante — saltamos título y encabezados
df_31_clean = df_31_clean.iloc[5:].copy()

# Renombramos la primera columna a 'territorio'
df_31_clean = df_31_clean.rename(columns={df_31_clean.columns[0]: 'territorio'})

# Eliminamos filas donde 'territorio' sea NaN — filas vacías separadoras
df_31_clean = df_31_clean.dropna(subset=['territorio']).copy()

# Convertimos columnas numéricas de texto a número
for col in ['Extra', 'Virgen', 'Lampante', 'TOTAL ']:
    df_31_clean[col] = pd.to_numeric(df_31_clean[col], errors='coerce')

# Eliminamos la nota legal — es la última fila
df_31_clean = df_31_clean.iloc[:-1].copy()

# Limpiamos el nombre residual del índice de columnas
df_31_clean.columns.name = None

# Reseteamos índice desde 0
df_31_clean = df_31_clean.reset_index(drop=True)

print(f"Dimensiones finales: {df_31_clean.shape}")
print("\nPrimeras 5 filas:")
print(df_31_clean.head())
print("\nÚltimas 3 filas:")
print(df_31_clean.tail(3))
```

```python
# Guardamos la tabla limpia como CSV en Google Drive
output_path = '/content/drive/MyDrive/proyecto_aceite_oliva/aceite_virgen_provincia_2023_clean.csv'

df_31_clean.to_csv(output_path, index=False)

print(f"✅ Archivo guardado en: {output_path}")
print(f"Filas: {df_31_clean.shape[0]} | Columnas: {df_31_clean.shape[1]}")
```

```python
# Limpieza tabla 7.12.1.5 - Serie histórica superficie, rendimiento, producción

df_15_clean = pd.read_excel(file_path, sheet_name='7.12.1.5', header=None)

# Filtramos solo filas con años válidos en columna 0
df_15_clean = df_15_clean[
    pd.to_numeric(df_15_clean[0], errors='coerce').between(2013, 2023)
].copy()

# Hay 3 bloques de 7 columnas cada uno: Almazara (1-7), Mesa (8-14), Doble aptitud (15-21)
# Construimos nombres de columna explícitos para cada bloque
cols_sufijo = ['superficie_total', 'superficie_produccion', 'superficie_recolectada',
               'arboles_diseminados_miles', 'rendimiento_qm_ha', 'rendimiento_kg_arbol',
               'produccion_miles_ton']

nombres = ['anio']
for prefijo in ['almazara', 'mesa', 'doble_aptitud']:
    nombres += [f'{prefijo}_{s}' for s in cols_sufijo]

# Seleccionamos columnas 0 a 21 (año + 3 bloques de 7)
df_15_clean = df_15_clean[list(range(0, 22))].copy()
df_15_clean.columns = nombres

# Los guiones "-" representan valores no disponibles — los convertimos a NaN
df_15_clean = df_15_clean.replace('-', pd.NA)

# Convertimos todas las columnas numéricas (todas menos 'anio')
for col in nombres[1:]:
    df_15_clean[col] = pd.to_numeric(df_15_clean[col], errors='coerce')

df_15_clean['anio'] = df_15_clean['anio'].astype(int)
df_15_clean = df_15_clean.reset_index(drop=True)

print(f"Dimensiones: {df_15_clean.shape}")
print(df_15_clean[['anio', 'almazara_produccion_miles_ton', 'mesa_produccion_miles_ton', 'doble_aptitud_produccion_miles_ton']])
```

```python
import pandas as pd
from google.cloud import bigquery

client = bigquery.Client(project=project_id)

# Diccionario: nombre de tabla → archivo CSV
tablas = {
    'precios_historicos': 'precios_historicos_clean.csv',
    'aceite_virgen_provincia_2023': 'aceite_virgen_provincia_2023_clean.csv',
    'destino_produccion_aceituna': 'destino_produccion_aceituna_clean.csv',
    'aceituna_almazara_productos': 'aceituna_almazara_productos_clean.csv',
    'superficie_rendimiento_produccion': 'superficie_rendimiento_produccion_clean.csv',
    'olivar_almazara_provincia': 'olivar_almazara_provincia_clean.csv'
}

drive_path = '/content/drive/MyDrive/proyecto_aceite_oliva/'

for tabla, archivo in tablas.items():
    df = pd.read_csv(drive_path + archivo)
    destination = f'{project_id}.{dataset_id}.{tabla}'
    job = client.load_table_from_dataframe(df, destination,
          job_config=bigquery.LoadJobConfig(write_disposition='WRITE_TRUNCATE'))
    job.result()
    print(f'✅ {tabla} — {df.shape[0]} filas cargadas')
```
